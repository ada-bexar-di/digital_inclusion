## rename GEO columns appropriately:

```python
def change_ages(df):
    #Drops Non-important GEOid
    #Renames GEO.id2 to blockgroup (useful for merging)
    colnames = {'GEO.id2':'blockgroup'}
    return df.drop(columns=['GEO.id','GEO.display-label']).rename(columns=colnames)
```

```python
ages = change_ages(ages)
```

```python
ages.head()
```

## removes Margin of Error fields (MOEs)

```python
def no_moes(df):
    return df[['blockgroup']+[col for col in df.columns[1:] if int(col[2:4]) == 1]]
```
